```python
import random

def gameA(stake, goal):

    bets = 0
    money = stake
    goal_amount = goal
    while money >= 0 and money < goal_amount:
        prob = random.randint(0,100)
    
        if prob > 50 and prob < 100:
            money = money + 1
            bets = bets + 1
            
        else:
            money = money - 1
            bets = bets + 1
            
    if money == goal_amount:
        win = True
    else:
        win = False
        
    return (bets, win)

    

def gameB(winProb, stake, goal):

    bets = 0
    money = stake
    goal_amount = goal
    while money > 0 and money < goal_amount:
        prob = random.randint(0,100)
    
        if prob <= winProb:
            money = money + 1
            bets = bets + 1
            
        elif prob > winProb:
            money = money - 1
            bets = bets + 1
    
    
    if money == goal_amount:
        win = True
    else:
        win = False
    
    return (bets, win)
        
    
# simulateA(num_of_games, money, goal_amount)
def simulateA(numTrials, stake, goal):

    bets_tot = 0
    num_wins = 0
    
    for i in range(0,numTrials):
        bets,win = gameA(stake, goal)
        bets_tot += bets
        if win == True:
            num_wins += 1
    print(int(bets_tot / numTrials), 'is the average number of bets made.')
    print((num_wins / numTrials), 'is the probability of winning.')


# simulateB(num_of_games, winProb, money, goal_amount)
def simulateB(numTrials, winProb, stake, goal):

    bets_tot = 0
    num_wins = 0
    
    for i in range(0,numTrials):
        bets,win = gameB(winProb, stake, goal)
        bets_tot += bets
        if win == True:
            num_wins += 1
    print(int(bets_tot / numTrials), 'is the average number of bets made.')
    print((num_wins / numTrials), 'is the probability of winning.')


def main():
    
    # Read input values from user
    # These will be passed through as parameters for each function
    initial_stake = int(input('Enter the initial stake:'))
    goal_amount = int(input('Enter the goal amount:'))
    num_of_games = int(input('How many times do you want to play?'))
    winProb = int(input('Enter a probability between 0 and 100:'))

    # Make Function calls
    gameA(initial_stake, goal_amount)
    gameB(winProb, initial_stake, goal_amount)
    simulateA(num_of_games, initial_stake, goal_amount)
    simulateB(num_of_games, winProb, initial_stake, goal_amount)

    # Display results

main()
```

```python
import numpy as np
import math

def LRiemannIntegral(f, a, b, n):
    #height f(x)
    #area = f(x) * delta_x
    area = 0
    delta_x = (b-a)/n
    x = a
    for k in range(n):
        area += f(x) * delta_x 
        x += delta_x
    return area

def f1(x):
    
    return (x * x)

def f2(x):
    
    return math.exp(-x**2)

def f3(x):
    
    return math.sin(x)

def f4(x):
   
    return (math.sin(x))/x

def main():
    
    n1 = 150000
    n2 = 1300
    n3 = 600
    n4 = 100000
    
    # Calling LRiemannIntegral function
    # Display results
    print(LRiemannIntegral(f1, 0, 10, n1))
    print(LRiemannIntegral(f2, 0, 10, n2))
    print(LRiemannIntegral(f3, 0, (2*math.pi), n3))
    print (LRiemannIntegral(f4, 2, 5, n4))

    
main()
```

```python
import numpy as np
def phi(n):
    #print('FIXME:  Implement phi(n)')
    original_number = n
    factors = []
    #For when the integer entered is even
    while n % 2 == 0:
        factors.append(2)
        n //= 2
    #For when the integer entered is odd
    nextprime = 3
    #Because if integer = 1 it will return number and divisor cannot be larger than number
    while n != 1 and nextprime <= n:
        if n % nextprime == 0:
            #if conditions are met, 3 will be added to list of factors
            factors.append(nextprime)
            #when the number you're dividing by does not give a remainder of 0
            n //= nextprime
            #update the divisor to the next possible prime factor
        else:
            nextprime = nextprime + 2
    uniquefactors = np.unique(factors)
    #print(uniquefactors)

    eulernum = original_number
    #Euler's Totient
    for i in uniquefactors:
        calc = 1 - (1/i)
        eulernum = eulernum * calc
    return int(eulernum)

def printTable():
    print('phi(n) for 1<=n<=100')
    print()
    print(' + ', end = '')
    for col in range(1,11):
        print(f'{col:4}', end = ' ')
    print()
    for row in range(0,100,10):
        print(f'{row:2}', end = ' ')
        for col in range(1,11):
            print(f'{phi(row+col):4}', end = ' ')  # corrected 10/3/2022
        print()
        
def demoProduct(p, k):
    #  Demonstrate the product formula for p^k
    p_k = p**k
    phi_p_k = p_k * (1 - (1/p))
    return phi_p_k


def main():
    printTable()
    print()
    p = int(input('Enter a prime number:'))
    k = int(input('Enter a positive integer:'))
    
    if int(demoProduct(p,k)) == phi(p**k):
        print(int(demoProduct(p,k)), phi(p**k))
    
    
main()
```
