1. Periodic table look up program.
Key -  atomic number, 2nd column in CSV file
Value - list containing name of elements (1st column) and atomic weight (4th column)

Prompt user to enter an atomic number then display the name of the element
Pose a question about the periodic table that you can answer with this data. Then write code to answer that question.

```python
import csv

Per_Table_csv = open('/Users/lilykraus/Downloads/elements.csv', 'r')
Per_Table_info = Per_Table_csv.read()

row_contents = Per_Table_info.split('\n')
#print(row_contents)

element_info_list = []
for i in range(len(row_contents)):
    element = row_contents[i]
    i = i +1
    element_info_list.append(element.split(','))

#Removing empty string
element_info_list.pop()

atomic_num_list = []
for i in range(len(element_info_list)):
    atomic_num = element_info_list[i][1]
    atomic_num_list.append(atomic_num)
    i = i + 1    
#print(len(atomic_num_list))

names_list= []
for i in range(len(element_info_list)):
    name = element_info_list[i][0]
    names_list.append(name)
    i = i + 1
weight_list = []
for i in range(len(element_info_list)):
    weight = element_info_list[i][3]
    weight_list.append(weight)
    i = i + 1

#Checking to make sure they're the same length
#print(names_list)
#print(weight_list)

names_weight_list = []
for i in range(len(element_info_list)):
    names_weight = names_list[i] +','+weight_list[i]
    names_weight = names_weight.split(',')
    #print(names_weight)
    names_weight_list.append(names_weight)
    i = i + 1
#print(names_weight_list)

Dict = {}
keys = atomic_num_list
values = names_weight_list

for i in range(len(keys)):
    Dict[keys[i]] = values[i]
#print(Dict)
#print(Dict.items())
user_num = input('Enter an atomic number:')

if user_num in Dict:
    print(Dict[user_num][0])
else:
    print('Sorry that is not a valid atomic number.')
print()


#Find elements that start with a certain letter
start_letter = input('Enter a letter:')
start_letter = start_letter.upper()

#Trying to get value[0] for each element in dictionary that has a name that starts with letter entered and turn it into a list
start_letter_elements = [value[0] for (key,value) in Dict.items() if start_letter in value[0]]

if len(start_letter_elements) == 0:
    print('There are no elements that start with', start_letter)

elif len(start_letter_elements) == 1:   
    print('The only element starting with', start_letter, 'is:')
    
    for i in range(len(start_letter_elements)):
        print(start_letter_elements[i],'\n')  
    
elif len(start_letter_elements) > 1:
    print('The elements beginning with', start_letter, 'are:')
    
    start_letters = ', '.join(start_letter_elements)
    print(start_letters)
    
```

iii. Compute and report the number of non-whitespace characters in the passage.

```python
import string

txt_file = open('/Users/lilykraus/Documents/HW4Test.txt', 'r')
file_contents = txt_file.read()
#print(file_contents)
file_contents2 = file_contents
#print(file_contents2)
file_contents = file_contents.strip('\n')
WordList = (file_contents.split())
WordList = [word.strip('),:(.]?!\"\'-\'[\'') for word in WordList]
WordList = [word.strip('1234567890') for word in WordList]

#remove empty string in wordlist
for word in WordList:
    if word == '':
        WordList.remove(word)

num_of_words = len(WordList)
print('Number of words:',num_of_words)

num_periods = file_contents.count('.')
#print(num_periods)
num_exclamation = file_contents.count('!')
#print(num_exclamation)
num_question = file_contents.count('?')
#print(num_question)

num_sentences = num_periods + num_exclamation + num_question
print('Number of sentences:',num_sentences)

#number of non whitespace characters

whitespace_char = list(string.whitespace)
#print(whitespace_char)

for i in range(len(whitespace_char)):
    file_contents2 = file_contents2.replace(whitespace_char[i], '')

#print(file_contents2)
num_non_whitespace = len(file_contents2)
print('Number of non-whitespace characters:', num_non_whitespace)
```
