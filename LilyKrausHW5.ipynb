```python
import math
import random

class Rectangle:
   
    def __init__(self, x=0, y=0, w=0, h=0, p=0):
    # Construct self with lower left corner (x, y), width w, and height h, oriented with x-y axes
        self.x = x
        self.y = y
        self.width = w
        self.height = h
        self.position = p
  
    def area(self):
    # Return the area of self.
        return self.width * self.height
    
    def perimeter(self):
    # Return the perimeter of self.
        return 2*(self.width + self.height)
    
    def diagonal(self):
    # Return the length of the diagonal of self.
    #diagonal of rectange = square root of length squared plus width squared
        dg = math.sqrt((self.height)**2 + (self.width)**2)
        return round(dg,2)
    
    def _containsPoint(self, p):  #provided
    # Helper function
    # Return True only if self contains the point p =(a, b)
        a, b = p
        a_in = (self.x <= a) and (a <= self.x + self.width)
        b_in = (self.y <= b) and (b <= self.y + self.height)
        return a_in and b_in
    
    def _corners(self): #provided
    # Helper function
    # Returns a list of the four corners of self, as 2-tuples
    #go over which corner is which
        return [(self.x, self.y), 
                (self.x + self.width, self.y), 
                (self.x + self.width, self.y + self.height),
                (self.x, self.y + self.height)]

    
    def intersects(self, other):
    # Return True if self intersects other, and False otherwise.
        co = other._corners()
        #for every point that is a corner in other, going through corners
        for p in co:
            #looking at all data together then getting one answer
            if self._containsPoint(p):
                return True
        #after for loop check to see if none of them are in 
            
        cs = self._corners()
        #go through corners
        for p in cs:
            #looking at all data together then getting one answer
            if other._containsPoint(p):
                return True
        #after for loop check to see if none of them are in
        return False

    def contains(self, other):
    # Return True if other is completely inside of self, and False otherwise.
        answer = True
        #for each point in other rectangle
        for p in other._corners():
            #if the point from other is not in self then return false
            if not self._containsPoint(p):
                return False
        return answer
        
      
    def __lt__(self, other):
    # Compare rectangles using the length of the diagonal
    # compare the diagonal lengths and then assign value of greatest length to variable
        diagonal_list  = []
        greatest_diagonal = 0
        
        for k in range(len(rectangle_list)):
            diagonal_list.append(rectangle_list[k].diagonal())
        #print(diagonal_list)
        greatest_diagonal = max(diagonal_list)
        greatest_rectangle = diagonal_list.index(greatest_diagonal)
        greatest_rectangle += 1
        return_value = 'Rectangle '+str(greatest_rectangle)+' is the largest rectangle with a diameter of '+str(greatest_diagonal)
        
        return return_value 
 
    
    def __str__(self):
    # Return a string representation of self.
        return f'Rectangle {self.position} at ({self.x}, {self.y}) with width {self.width} and height {self.height}'
        
        
if __name__ == '__main__':    

# Create a list of 10 rectangle with random widths, heights, x- and y- locations, 
# with all numbers in the float interval (0, 10)
    rectangle_list = []
    for k in range(0,10):
        rand_x = round(random.uniform(0,10),2)
        rand_y = round(random.uniform(0,10),2)
        rand_w = round(random.uniform(0,10),2)
        rand_h = round(random.uniform(0,10),2)
        p = k+1
        r = Rectangle(rand_x,rand_y,rand_w,rand_h,p)
        rectangle_list.append(r)
    
        
    # Display the 10 rectangles using print and a for loop
    
    for k in range(len(rectangle_list)):
        print(rectangle_list[k])
    print()
    
    # Display the diagonal lengths of each of the 10 rectangles
    
    for k in range(len(rectangle_list)):
        print('Rectangle',k+1, 'has a diagonal length of', rectangle_list[k].diagonal())
    print()
    
    # Determine if any of your 10 rectangles is contained in another of the 10
    # Print your findings
    contains_list = []
    for k in range(len(rectangle_list)):
        #going through list of rectangles and comparing one to the rest of the rectangles
        for m in range(len(rectangle_list)):
            if rectangle_list[k].contains(rectangle_list[m]) == True and rectangle_list[k] != rectangle_list[m]:
                contains_list.append(rectangle_list[k])
                #print(rectangle_list[k],k, 'contains', rectangle_list[m],m)
                print('Rectangle', k+1, 'contains', 'Rectangle', m+1)
    if len(contains_list) == 0:
        print('None of the rectangles contain one of the other rectangles')

        
    print()
    # Determine if any of your 10 rectangles intersects another of the 10
    # Print your findings
    
    #make another list to store rectangle info
    intersect_rectangles = []
    for k in range(len(rectangle_list)):
        my_intersections = []
        #going through list of rectangles and comparing one to the rest of the rectangles
        for m in range(len(rectangle_list)):
            #exclude same and repeat rectangles
            if rectangle_list[k] != rectangle_list[m]:
                #putting rectangles in separate list so that they aren't repeated
                if rectangle_list[k].intersects(rectangle_list[m]) == True: 
                    my_intersections.append(m+1)
  
        intersect_rectangles.append(my_intersections)
        
    for i in range(len(intersect_rectangles)):
        if len(intersect_rectangles[i]) == 0:
            print('Rectangle', i+1, 'does not intersect any other rectangles')
        else:
            print('Rectangle',i+1, 'intersects rectangles', intersect_rectangles[i])
                    #print('Rectangle',k+1, 'intersects', 'Rectangle', m+1)

    print()
    
    # Determine the "largest" of your 10 rectangles, using the __lt__ function you defined 
    print(rectangle_list[k].__lt__(rectangle_list[0]))
```

```python
import csv

# Periodic Table with 103 elements - one line string
#for molecular weight take in different atomic weights and add them

class Element:
    def __init__(self, an=1, n='Hydrogen', abr = 'H', aw = 1.01):
        self.atomic_number = an
        self.name = n
        self.abrev = abr
        self.atomic_weight = aw
        
    def __str__(self):
        return f'{self.name}, {self.abrev}, Atomic Number:{self.atomic_number}, Atomic weight:{self.atomic_weight}'
        
    def __lt__(self, other):
        return self.atomic_number < other.atomic_number
       
lowercase = 'abcdefghijklmnopqrstuvwxyz'
digits = '0123456789'

# Populate the element_dict using the data in filename
# Dictionary keys should be abbreviations
class PeriodicTable:
    def __init__(self, filename):
        
        Per_Table_csv = open('/Users/lilykraus/Downloads/elements.csv', 'r')
        Per_Table_info = Per_Table_csv.read()

        row_contents = Per_Table_info.split('\n')
        #print(row_contents)

        element_info_list = []
        for i in range(len(row_contents)):
            element = row_contents[i]
            i = i +1
            element_info_list.append(element.split(','))
        #print(element_info_list)
        #Removing empty string
        element_info_list.pop()

        self.element_Dict = {item[2]:Element(item[1],item[0],item[2], float(item[3])) for item in element_info_list}
        #print(self.element_Dict['O'])
        
    def __str__(self):
        return f'Periodic table with {len(self.element_Dict)} elements' 
    
        
    def addElement(self, anElement):
        # Add the element anElement to the periodic table if it does not already exist in the table
        #print(len(self.element_Dict))
        if anElement.abrev not in self.element_Dict:
            self.element_Dict[anElement.abrev] = anElement
            #print(len(self.element_Dict))
            return 'Element added to dictionary'
        else:
            return 'Already in dictionary'
                
    def _nextAtom(mstring): 
        #provided
        # Helper method to find the next atom and number in a string representing a molecular formula
        # Returns one element name, one number and the remainder of the string with this info removed
        mstring = mstring.replace(' ', '')
        k = 0
        num = 1
        numstr = ''
        abr = ''
        if k < len(mstring):
            abr = mstring[k]
            if (k + 1 < len(mstring)) and mstring[k+1] in lowercase:
                abr += mstring[k+1]
                k += 2
            else:
                k +=1
            if k < len(mstring) and mstring[k] in digits:
                while k < len(mstring) and mstring[k] in digits :
                    numstr += mstring[k]
                    k+=1
                num = int(numstr)
    
        return (abr, num, mstring[k:])
        
    def molecular_weight(self, mstr):
        # Return a number representing the weight of the molecule given in the string mstr        
        answer = 0
        while len(mstr) > 0:
            a,b,mstr = PeriodicTable._nextAtom(mstr)
            #use a as key
            weight = self.element_Dict[a].atomic_weight
            #multiply a and b then add to answer
            tot_weight = weight * b
            answer = answer + tot_weight
               
        return f'Molecular weight: {answer}'
        
if __name__ == '__main__':            
    
    my_table = PeriodicTable('elements.csv')
    anElement = Element(112,'Copernicium','Cn', 173)
    print('Before:')
    print(my_table)
    print(my_table.addElement(anElement))
    print()
    print('After:')
    print(my_table)
    print('Added entry:',my_table.element_Dict['Cn'])
    print(my_table.addElement(anElement))
    print(my_table)
    print()
    
    print(my_table.molecular_weight('H75O2C450'))  #5512.25
    print(my_table.molecular_weight('He25Li8La12'))  #1822.44
    print(my_table.molecular_weight('H2O'))  #18.02
      
```
