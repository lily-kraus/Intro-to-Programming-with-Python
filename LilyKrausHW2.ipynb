2.  (10 points)  Factoring positive integers

```python
# Your code for the factoring problem goes here
#n%f == 0 means f is a factor of n
integer = int(input('Enter an integer to be factored:'))
original_number = integer
factors = []
#For when the integer entered is even
while integer % 2 == 0:
    factors.append(2)
    integer //= 2

#How to test if number is prime? 


#For when the integer entered is odd
nextprime = 3
#Because if integer = 1 it will return number and divisor cannot be larger than number
while integer != 1 and nextprime <= integer:
    if integer % nextprime == 0:
        #if conditions are met, 3 will be added to list of factors
        factors.append(nextprime)
        #when the number you're dividing by does not give a remainder of 0
        integer//= nextprime
        #update the divisor to the next possible prime factor
    else:
        nextprime = nextprime + 2 

#How to test if number is prime?     
#if the length of factors is 1, print that number is prime.

if len(factors) == 1:
    print(original_number, 'is a prime number.')

#for i in range(len(factors)):
#    print(original_number, '=', )
    
print(original_number, '=',' * '.join(map(str, factors)))
```

3.  Montecarlo Approximation for Pi, with visualization

```python
# Your code for the pi approximation problem goes here

#probability that a dart will fall inside the circle = pi/4
#probability that a dart will fall inside the circle = num darts in circle / total number of darts thrown
#pi/4 = num darts in circle / total num of darts thrown
#pi = 4 * num  darts in circle / total num of darts thrown
#area shaded / total area = numdarts inside / total num darts-3

import math
import random

numdarts = input('How many darts should I throw?')
numdarts = int(numdarts)

#create empty list to hold points as you go
insidepoints = []
outsidepoints = []

while numdarts > 0:
    #get random x and y coordinates in range 0 to 1
    for i in range (0,numdarts):
        random.seed()
        x = random.uniform(0,1)
        y = random.uniform(0,1)

        points = (x,y)

        r = (x**2 + y**2)
        
        if r < 1: 
            insidepoints.append(points)
        elif r > 1:
            outsidepoints.append(points)
    
    #print(insidepoints)
   # print(outsidepoints)
    numdarts = input('How many darts should I throw? Enter a negative number to stop.')
    numdarts = int(numdarts)
    
numoutsidepoints = len(outsidepoints)
numinsidepoints = len(insidepoints)
totalpoints = len(insidepoints) + len(outsidepoints)

piApproximation = (numinsidepoints / totalpoints) * 4
print('Approximation of pi:',piApproximation)

#piApproximation = insidepointslength / numdarts
#print(piApproximation)
    
#pi approximation here   
    
    
import matplotlib.pyplot as plt
import numpy as np  # numpy also has lots of built in math functions

plt.figure(figsize = (6, 6))

for p in insidepoints:
    plt.plot(p[0], p[1], color = 'green', marker = 'o')

for p in outsidepoints:
    plt.plot(p[0], p[1], color = 'red', marker = 'o')

```
